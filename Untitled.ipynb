```python
def filter_cars(cars, param, kriterui):
    filter_car = []
    for car in cars:
        for car_info in car:
            if param in car[car_info]:
                if car[car_info][param] == kriterui:
                    filter_car.append(car_info)

    return filter_car


filter_car = filter_cars(cars, "выпуск", 2018)
pprint(filter_car)
```
